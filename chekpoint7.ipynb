```python
import pandas as pd
import numpy as np
sheet = pd.read_csv(r"C:\Users\hp\Desktop\python\data-preprocessing titanic.csv")
df=pd.DataFrame(sheet)
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Decision Tree Prediction

#importing relevant libraries
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

#features extraction
x=df.drop(["Survived",'Title'], axis=1)
y= df["Survived"]

#splitting data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)
```

```python
#applying tree algorithm
tree = tree.DecisionTreeClassifier()  
tree.fit(x_train, y_train)   #fitting our model
y_pred=tree.predict(x_test)   # evaluating our model
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
import graphviz
from sklearn.tree import export_graphviz

dot_data = export_graphviz(tree, out_file=None) 
graph = graphviz.Source(dot_data)
graph.render("df")
graph
```

```python
tree = DecisionTreeClassifier(criterion='entropy',splitter='random') 
tree.fit(x_train, y_train)   #fitting our model
y_pred=tree.predict(x_test)   # evaluating our model
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
import graphviz
from sklearn.tree import export_graphviz

dot_data = export_graphviz(tree, out_file=None) 
graph = graphviz.Source(dot_data)
graph.render("df")
graph
```

```python
tree = DecisionTreeClassifier(min_samples_leaf=3,max_depth=3,max_leaf_nodes=5) 
tree.fit(x_train, y_train)   #fitting our model
y_pred=tree.predict(x_test)   # evaluating our model
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
import graphviz
from sklearn.tree import export_graphviz

dot_data = export_graphviz(tree, out_file=None) 
graph = graphviz.Source(dot_data)
graph.render("df")
graph
```

```python
###Random Forest Code
###Random Forest prediction
for i in range (10,200,10):
    from sklearn.ensemble import RandomForestClassifier #Importing Random Forest Classifier
    from sklearn import metrics  # Importing metrics to test accuracy
    clf=RandomForestClassifier(n_estimators=i)  #Creating a random forest with 100 decision trees
    clf.fit(x_train, y_train)  #Training our model
    y_pred=clf.predict(x_test)  #testing our model
    print("Accuracy:", metrics.accuracy_score(y_test, y_pred))  #Measuring the accuracy of our model
```
